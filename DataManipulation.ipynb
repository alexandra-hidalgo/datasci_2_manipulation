## Creating new columns

```python
# Generate a fake healthcare dataset

fake = Faker()
num_patients = 100

data = {
    'Blood_Sugar': [fake.random_int(min=50, max=200) for _ in range(num_patients)],
    'Blood_Pressure': [random.uniform(90, 180) for _ in range(num_patients)],
}

df = pd.DataFrame(data)

# Define clinically defined thresholds for categorization
threshold_high = 100

# Create a new column "Blood_Pressure_Status" based on thresholds
df['Blood_Sugar_Status'] = df['Blood_Sugar'].apply(lambda x: 'High' if x >= threshold_high else 'Normal')

print(df[['Blood_Sugar', 'Blood_Pressure', 'Blood_Sugar_Status']].head(15))
```
